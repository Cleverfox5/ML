## 2) Удалить выбросы и дубли, обработать пропущенные значения

### Выбросы

В предыдущей работе была написана функция для визуализации первого выброса, изменим таки образом, чтобы она удалала все выбросы.

```python
def delete_blowouts(df, column):
    # вычисляем первый квартиль, третий, межквартальный размах, на основе их - верхнюю и нижнюю границу
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    upper_boundary = Q3 + (1.5 * IQR)
    lower_boundary = Q1 - (1.5 * IQR)

    # Если вышел из диапозона - выброс
    outliers = df[column][(df[column] > upper_boundary) | (df[column] < lower_boundary)]
    # удаление выбрасов
    print(f'Количество элементов до удаления выбросов: {df["Price"].count()}')
    df.drop(outliers.keys(), inplace=True)
    print(f'После: {df["Price"].count()}')
```

## 3) Применение метода KMeans

Оставим только числовые атрибуты

```python
df = df.select_dtypes(include=['float64', 'int64'])
#df = only_numbers_df.drop(columns=['Unnamed: 0'])
df.head()
```

Функция для определения оптимального количества кластеров (метод локтя)

```python
def number_of_clusters(df, X1):
    inertia = []
    for n in range(1 , 11):
        algorithm = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=300, 
                            tol=0.0001,  random_state= 111  , algorithm='lloyd') )
        algorithm.fit(X1)
        inertia.append(algorithm.inertia_)
    
    plt.figure(1, figsize=(15, 6))
    plt.plot(np.arange(1, 11), inertia, 'o')
    plt.plot(np.arange(1, 11), inertia, '-', alpha=0.5)
    plt.xlabel('Количество кластеров')
    plt.ylabel('Инерция')
    plt.show()
```

Функция, разделяющая данные на группы (K-Means)

```python
def group_values(df, X1, clusters_num, first_col_name, second_col_name):
    algorithm = (KMeans(n_clusters = clusters_num ,init='k-means++', n_init = 10 ,max_iter=300, 
                            tol=0.0001,  random_state= 111  , algorithm='elkan') )
    algorithm.fit(X1)
    labels1 = algorithm.labels_
    centroids1 = algorithm.cluster_centers_

    h = 0.02
    x_min, x_max = X1[:, 0].min() - 1, X1[:, 0].max() + 1
    y_min, y_max = X1[:, 1].min() - 1, X1[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    Z = algorithm.predict(np.c_[xx.ravel(), yy.ravel()])

    plt.figure(1, figsize=(15, 7))
    plt.clf()
    Z = Z.reshape(xx.shape)
    plt.imshow(Z, interpolation='nearest', 
               extent=(xx.min(), xx.max(), yy.min(), yy.max()),
               cmap=plt.cm.Pastel2, aspect='auto', origin='lower')
    plt.scatter( x = first_col_name ,y = second_col_name , data = df , c = labels1 , 
                s = 200 )
    plt.scatter(x = centroids1[: , 0] , y =  centroids1[: , 1] , s = 300 , c = 'red' , alpha = 0.5)
    plt.ylabel(second_col_name) , plt.xlabel(first_col_name)
    plt.show()
```

#### Проверим разделение на классы на примере связи - год и длинна

```python
X1 = df[['Length' , 'Year']].iloc[: , :].values
number_of_clusters(df, X1)
```

Из графика видно, что для данного отношения лучше всего подоёдёт разделение на 2 кластера

```python
group_values(df, X1, 2, 'Length', 'Year')
```
